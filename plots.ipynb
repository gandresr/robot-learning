```python
import pickle, os
import matplotlib.pyplot as plt
```

#### Define Results Folder

```python
RESULTS_DIR = 'results'
```

```python
# prediction variance results
with open(os.path.join(RESULTS_DIR, "lift_ph_scss_ensemble_5_random.txt"), "rb") as fp:
    lift_ph_random_scss_rate = pickle.load(fp)
x_lift_ph_random = [1000]
for ele in lift_ph_random_scss_rate[1:]:
    x_lift_ph_random.append(300 + x_lift_ph_random[-1])
# print(lift_ph_random_scss_rate)
plt.plot(x_lift_ph_random, lift_ph_random_scss_rate, label = 'Random select, Lift_ph')

with open(os.path.join(RESULTS_DIR, "lift_ph_scss_ensemble_5_variance.txt"), "rb") as fp:
    lift_ph_var_scss_rate = pickle.load(fp)
x_lift_ph_var = [1000]
for ele in lift_ph_var_scss_rate[1:]:
    x_lift_ph_var.append(300 + x_lift_ph_var[-1])
# print(lift_ph_var_scss_rate)
plt.plot(x_lift_ph_var, lift_ph_var_scss_rate, label = 'Prediction Variance, Lift_ph')

with open(os.path.join(RESULTS_DIR, "lift_mh_scss_ensemble_5_variance.txt"), "rb") as fp:
    lift_mh_var_scss_rate = pickle.load(fp)
lift_mh_var_scss_rate = lift_mh_var_scss_rate[2:]
x_lift_mh_var = [1000]
for ele in lift_mh_var_scss_rate[1:]:
    x_lift_mh_var.append(300 + x_lift_mh_var[-1])
# print(lift_mh_var_scss_rate)
plt.plot(x_lift_mh_var, lift_mh_var_scss_rate, label = 'Prediction Variance, Lift_mh')

with open(os.path.join(RESULTS_DIR, "lift_mh_scss_ensemble_5_random.txt"), "rb") as fp:
    lift_mh_random_scss_rate = pickle.load(fp)
lift_mh_random_scss_rate = lift_mh_random_scss_rate[2:]
x_lift_mh_random = [1000]
for ele in lift_mh_random_scss_rate[1:]:
    x_lift_mh_random.append(300 + x_lift_mh_random[-1])
# print(lift_mh_random_scss_rate)
plt.plot(x_lift_mh_random, lift_mh_random_scss_rate, label = 'Random Selection, Lift_mh')

with open(os.path.join(RESULTS_DIR, "can_ph_scss_ensemble_5_random.txt"), "rb") as fp:
    can_ph_random_scss_rate = pickle.load(fp)
can_ph_random_scss_rate = can_ph_random_scss_rate[2:]
x_can_ph_random = [1000]
for ele in can_ph_random_scss_rate[1:]:
    x_can_ph_random.append(300 + x_can_ph_random[-1])
# print(can_ph_random_scss_rate)
plt.plot(x_can_ph_random, can_ph_random_scss_rate, label = 'Random Selection, Can_ph')

with open(os.path.join(RESULTS_DIR, "can_ph_scss_ensemble_5_variance.txt"), "rb") as fp:
    can_ph_var_scss_rate = pickle.load(fp)
can_ph_var_scss_rate = can_ph_var_scss_rate[2:]
x_can_ph_var = [200]
for ele in can_ph_var_scss_rate[1:]:
    x_can_ph_var.append(300 + x_can_ph_var[-1])
# print(can_ph_var_scss_rate)
plt.plot(x_can_ph_var, can_ph_var_scss_rate, label = 'Prediction Variance, Can_ph')

# add can_mh with init and from 1000 samples
plt.legend()
plt.show()
```
